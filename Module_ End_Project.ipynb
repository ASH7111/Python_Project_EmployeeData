## `5) Find if there is any correlation between age and salary, represent it visually.`

```python
## (Plotting the Age in X-axis & Salary in Y-axis)

x= df['Age']
y= df['Salary']

plt.scatter(x, y)

plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')

plt.show()
```

```python
## Here, we can oberve that,
'''
o From the Scatter Plot we can see that, there is no such correlation found between 'Salary and Age'.
  
o We can see that, for the age group 30 to 35, salary is least for more majority of people.

o In all the age groups, there are employees with minimum salary and they are the majority of the employees.

o So we can conclude that, not only 'Age' can determine the 'Salary', other factors are also considered.

'''
```
